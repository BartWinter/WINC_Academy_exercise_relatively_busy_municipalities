```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base_url = 'https://github.com/WincAcademy/practice_data/raw/main/data/geo/'
provinces_url = f"{base_url}provinces.csv"
province_names = ["groningen","friesland","drenthe","overijssel","flevoland","gelderland","utrecht","noord-holland","zuid-holland","zeeland","noord-brabant","limburg"]
municipalities_url = [f"{base_url}municipalities/{prov_name}.csv" for prov_name in province_names]

prov_df = pd.read_csv(provinces_url)
#display(prov_df)

#concatenate the CSV files for the municipalities

mun_df = pd.concat(pd.read_csv(url) for url in municipalities_url)
display(mun_df)
```

```python
#Drop columns that are irrelevant for municipalities
mun_df2 = mun_df.drop(columns=['CBS code'])

#Rename columns
mun_df3 = mun_df2.rename(columns={'Surface Area in sq km':'surface', 'Province':'province','Inhabitants':'inhabitants','Municipality':'municipality'})

#Calculate new column
mun_df3['province_density'] = mun_df3['inhabitants']/mun_df3['surface']

mun_df4 = mun_df3.rename(columns={'province_density':'mun_density'})
display(mun_df4)
```

**Calculate the busiest and the least busy municipality per province using groupby**

```python
mun_df4.groupby(['province'], as_index=False)[['municipality','mun_density']].agg(['max', 'min'])
```
